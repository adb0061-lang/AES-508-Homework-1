```python
import folium
import pandas as pd
```

```python
df = pd.read_csv('HW1_Data - Sheet1.csv')
print(df)
```

```python
m = folium.Map(location=[df["Lat"].mean(), df["Long"].mean()], zoom_start=10)

for _, row in df.iterrows():
    folium.Marker(
        location=[row["Lat"], row["Long"]],
        tooltip="Click me!",
        popup=row["Location"],
        icon=folium.Icon(icon="cloud"),
    ).add_to(m)

m
```
